```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.broadwayworld.com/grosses.cfm')
```

```python
shows = driver.find_elements_by_xpath('//*[@id="content"]/div/div/div/div/div/div/div/table/tbody/tr/td/table[2]/tbody/tr')

rows = []

for show in shows[1:-2]:
    print('--------')
    info = show.find_elements_by_tag_name('td')
    
    row = {}
    
    title = info[0]
    row['title'] = title.text
    
    current_week = info[1]
    row['current_week'] = current_week.text
    
    previous_week = info[2]
    row['previous_week'] = previous_week.text
    
    difference = info[3]
    row['difference'] = difference.text
    
    avg_ticket = info[6]
    row['avg_ticket'] = avg_ticket.text
    
    top_ticket = info[7]
    row['top_ticket'] = top_ticket.text
    
    seats_sold = info[8]
    row['seats_sold'] = seats_sold.text
    
    seats_available = info[9]
    row['seats_available'] = seats_available.text
    
    pct_sold = info[11]
    row['pct_sold'] = pct_sold.text
    
    pct_sold_previous = info[12]
    row['pct_sold_previous'] = pct_sold_previous.text
    
    diff_sold = info[13]
    row['diff_sold'] = diff_sold.text
    
    print(row)
    rows.append(row)
```

```python
df_2207 = pd.DataFrame(rows)
df_2207
```

```python
df_2207['avg_ticket'] = df_2207['avg_ticket'].str.extract(r"(\d+.\d+)")
#df_2207['avg_ticket']
df_2207
```

```python
df_2207['current_week'] = df_2207['current_week'].str.extract(r"((\d+,.*))")
df_2207['previous_week'] = df_2207['previous_week'].str.extract(r"((\d+,.*))")
df_2207['difference'] = df_2207['difference'].str.extract(r"((\d+,.*))")
df_2207.head()
```

```python
df_2207['diff_sold'] = df_2207['diff_sold'].str.extract(r"(.*)%")
df_2207['pct_sold'] = df_2207['pct_sold'].str.extract(r"(.*)%")
df_2207['pct_sold_previous'] = df_2207['pct_sold_previous'].str.extract(r"(.*)%")
df_2207.head()
```

```python
df_2207['top_ticket'] = df_2207['top_ticket'].str.extract(r"(\d+)")
df_2207.head()
```

```python
df_2207['title'] = df_2207['title'].str.replace("\n","$")
df_2207.head()
```

```python
df_2207['title'] = df_2207['title'].str.extract(r"(.*)\$")
df_2207.head()
```

```python
df_2207
```

```python
df_2207 = df_2207.dropna(thresh=3)
df_2207.head(1)
```

```python
df_2207['current_week'] = df_2207['current_week'].str.replace(',','')
df_2207['difference'] = df_2207['difference'].str.replace(',','')
df_2207['previous_week'] = df_2207['previous_week'].str.replace(',','')
df_2207['seats_available'] = df_2207['seats_available'].str.replace(',','')
df_2207['seats_sold'] = df_2207['seats_sold'].str.replace(',','')
df_2207.head(1)
```

```python
df_2207['avg_ticket'] = df_2207['avg_ticket'].astype(float)
df_2207['diff_sold'] = df_2207['diff_sold'].astype(float)
df_2207['pct_sold'] = df_2207['pct_sold'].astype(float)
df_2207['pct_sold_previous'] = df_2207['pct_sold_previous'].astype(float)
df_2207['current_week'] = df_2207['current_week'].astype(int)
df_2207['difference'] = df_2207['difference'].astype(int)
#df_2207['previous_week'] = df_2207['previous_week'].astype(int)
df_2207['seats_available'] = df_2207['seats_available'].astype(int)
df_2207['seats_sold'] = df_2207['seats_sold'].astype(int)
df_2207['top_ticket'] = df_2207['top_ticket'].astype(int)

df_2207.dtypes
```

```python
#df_2207.previous_week
df_2207['previous_week'] = df_2207.previous_week.fillna(0)
df_2207['previous_week'] = df_2207.previous_week.astype(int)
df_2207.head()
```
